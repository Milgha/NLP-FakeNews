```python
def gettweettext(tweet_id_lists):
    # This function should take a list of tweet ids, get the text from Twitter
    # and save the resulting text as a tuple of tweet id, tweet text.
    # Make sure to create a tweepy Client instance called 'client' to use.

    output_list = []
    for i in tweet_id_lists:
        tweet = client.get_tweet(i)
        tweet_id = i
        tweet_tuple = (tweet_id, str(tweet.data)) 
        output_list.append(tweet_tuple)
    return output_list
```

```python
def gettweets(saved_df,hours):
    '''
    This function will iterate over 5 minutes time periods and fetch 100 tweets and append the dataframe,
    then save it to csv file for the overal hours specified in the function.
    
    '''    
    
    i = hours*12
    while i > 0:
        print("Extracting tweets...")
        saved_df = pd.read_csv('tweets.csv')
        tweets = client.search_recent_tweets(query = query, max_results=100, since_id=saved_df.id.iloc[-1],)
        print("100 tweets extracted" )
        df= pd.DataFrame(tweets[0])
        df.drop(columns='edit_history_tweet_ids', axis=1, inplace=True)
        tweets_df = saved_df.append(df)
        tweets_df.to_csv('Data/tweets.csv', index=False)
        i -= 1
        print(tweets_df.info())
        print(f'Extracted tweets: {len(tweets_df)}' )
        time.sleep(60*5.15)
    return "Done!"
```
